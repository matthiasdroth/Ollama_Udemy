## 1. Text

```python
from ollama import Client
ollama_client  = Client(host="http://localhost:11434")
output = ollama_client.chat(model="llama3.2", messages=[
    {
        "role": "user",
        "content": "What is the capital of the UK?"
    }
])
print(output["message"]["content"])
```

## 2. Text and images
The ***image analysis*** may take several minutes.

```python
%%time
output = ollama_client.chat(model='llama3.2-vision', messages=[
    {
        "role": "user",
        "content": "Please analyze the following image and describe its content in detail.",
        "images" : ["/home/matthias/Desktop/MachineLearning/Ollama_Udemy/Images/Ollama_cli_commands.png"]
        # Tip: Look the image's absolute (!) path up directly in the file system.
    }
])
print(output["message"]["content"])
```
