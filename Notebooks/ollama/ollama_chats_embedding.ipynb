```python
from ollama import Client
ollama_client = Client(host="http://localhost:11434")
output = ollama_client.embed(model="llama3.2", input=["dog", "cat"])
print(output["embeddings"])
```

```python
print(output["embeddings"][0]) # embedding for "dog"
print(output["embeddings"][1]) # embedding for "cat"
```

```python
print(len(output["embeddings"][0]))  # dimension of the embedding vector for "dog"
print(len(output["embeddings"][1]))  # dimension of the embedding vector for "cat"
```
