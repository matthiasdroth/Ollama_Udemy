# Running Ollama via OpenAI's API Integration
Run this within the virtual environment **(env_ollama)**!

```python
!which python
```

An **api key is not required** for this integration since the `client` is running on local hardware.

```python
from openai import OpenAI
client = OpenAI(base_url="http://localhost:11434/v1", api_key="")
completion = client.chat.completions.create(
    model="llama3.2",
    store=True,
    messages=[
        {"role": "user", "content": "Write a haiku about AI!"}
    ]
)
print(completion.choices[0].message.content)
```
