```python
from ollama import Client
ollama_client = Client()
output = ollama_client.chat(model="llama3.2", messages=[
    {
        "role": "user",
        "content": "What is the capital of USA and list me states in the USA!"
    }
], stream=True)
for chunk in output:
    print(chunk["message"]["content"], end="", flush=True)
```
